```
from ultralytics import YOLO
import torch


# model = YOLO("yolov8n.pt")
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# results = model("https://ultralytics.com/images/bus.jpg")
# results = model.predict(source="0", show=True)

results = model('/content/20.jpg')

results.print()
results.show()

results_info = results.xyxy[0]
# results.pandas().xyxy[0]

# Top-Left corner coordinates
xmin = results_info[0][0]
ymin = results_info[0][1]

# Bot-Right corner coordinates
xmax = results_info[0][2]
ymax = results_info[0][3]

# Confidence and detected class
confidence = results_info[0][4]
obj_class = results_info[0][5]

# Calculate Bounding box size

Height = ymax - ymin
Width = xmax - xmin

# print(xmin)

print('Height: ',Height)
print('Width: ',Width)
```
